## Transfer training of DeepMoji 
***

```python
DATASET_PATH = 'DeepMoji/data/PsychExp/raw.pickle'
nb_classes = 7

with open('DeepMoji/model/vocabulary.json', 'r') as f:
    vocab = json.load(f)

# Load dataset.
data = load_benchmark(DATASET_PATH, vocab, extend_with=10000)

# Reduce dimension to 1 d for finetunning
one_data=list()
for label in data['labels']:
    one_data.append(label.argmax(axis=1))

# Set up model and finetune
model = deepmoji_transfer(7, data['maxlen'], PRETRAINED_PATH,extend_embedding=data['added'])
model.summary()

model, acc = finetune(model, data['texts'], one_data, 7, data['batch_size'], metric='acc', method='chain-thaw',verbose=3)
#model, f1 = class_avg_finetune(model, data['texts'], data['labels'],
#                               nb_classes, data['batch_size'], method='last')
```

```python
#################################################### TESY #############################################
maxlen = 40
st = SentenceTokenizer(vocab, maxlen)
tokenized, _, _ = st.tokenize_sentences([u'Hello world',u'Fuck you, bitch', u'A beautiful girl', u'ni hao'])
predict_list=list()
A=model.predict(tokenized[0:2])
predict_list.append(A)
#print (A)
B=model.predict(tokenized[2:4])
predict_list.append(B)
print (np.concatenate(predict_list, axis=0))
```

```python
def predict(chunk_df):
  
    #longtxt_df = chunk_df[chunk_df['txt'].str.count(',')>2]
    #shorttxt_df = chunk_df[chunk_df['txt'].str.count(',')<=2]
    word_list=chunk_df['txt'].tolist()
    sentence_list=[re.sub(', https[^\,\]]*',',',txt) for txt in word_list]
    sentence_list=[".".join(re.findall('\[(?:[^\]|]*\|)?([^\]|]*)\]', txt)).replace(',', '') for txt in sentence_list]
    
    maxlen = 40
    st = SentenceTokenizer(vocab, maxlen)
    tokenized, _, _ = st.tokenize_sentences(sentence_list)
    
    return model.predict(tokenized)
```

***
## Chunkize Tweet txt to do prediction
***

```python
n=0
predict_list=list()
chunk=100000
print (year, " Chunck size(lines): ",chunk)
for index, chunk_df in enumerate(pd.read_csv('data/US_'+year+'_integrated.csv',encoding='utf-8',usecols=['txt'],header=0,chunksize=chunk)):
    time1 = time.time()
    results=predict(chunk_df)
    #predict_list.append(results)
    n=n+chunk
    time2 = time.time()
    print ("First ", n, ". Execution time: ", (time2-time1), 's Lengh: ',len(predict_list))
    np.savetxt(year+'_'+str(index)+'_emo.out', results, delimiter=',') 
print ("Execution Done")
#prob=np.concatenate(predict_list,axis=0)
#np.savetxt(year+'_emo.out', prob, delimiter=',') 
```
